## Inference using chat completion API endpoint

Utilize the `v1/chat/completions` to obtain both the `reason` and the `content` fields from the model output. <br> 

Refer to the [sync API endpoints](https://docs.baseten.co/inference/calling-your-model#sync-api-endpoints) to obtain the correct endpoint URL. <br>
A typical sync endpoint should be following the format of `https://model-{model-id}.api.baseten.co/environments/{production}/sync/{route}`, where <br>
* `model_id` refers to the model ID from the Baseten deployment
* `production` refers to the deployment type or ID
* `route` refers to the routes in the vLLM server, which is `v1/chat/completions` in this deployment

See [Reference Documentation](https://docs.baseten.co/reference/inference-api/overview#deployed-model-endpoints) for more information.

```python
import requests
from IPython.display import display, Markdown

ENDPOINT_URL = "" # Replace with your endpoint URL. Example: https://model-{model_id}.api.baseten.co/environments/production/sync/v1/chat/completions
API_KEY = "" # Replace with your API key
assert ENDPOINT_URL, "Set endpoint_name above before running inference"

# Reasoning models produce more tokens due to chain-of-thought, so set a higher limit
MAX_TOKENS = 8192
TEMPERATURE = 0.3
```

```python
def chat_generation(user_prompt):
    data = {
        'messages': [{"role": "user", "content": user_prompt}],
        'temperature': TEMPERATURE,
        'max_tokens': MAX_TOKENS
        }

    response = requests.post(
        ENDPOINT_URL,
        headers={"Authorization": f"Api-Key {API_KEY}"},
        json=data,
    )
    output = response.json()
    message = output["choices"][0]["message"]

    reasoning = message.get("reasoning_content") or message.get("reasoning", "")
    answer = message.get("content", "")
    return reasoning, answer
```

```python
user_prompt = "What is cybersecurity?"

reasoning, answer = chat_generation(user_prompt)

print("=== Reasoning (chain-of-thought) ===")
display(Markdown(reasoning))
print()
print("=== Answer ===")
display(Markdown(answer))
```
